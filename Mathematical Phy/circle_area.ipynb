```python
import numpy as np
import random
import matplotlib.pyplot as plt

x = []
y = []
for i in range(10000):
    x.append(random.random())
    y.append(random.random())
d = []
c = []
for i in range(len(x)):
    d_i = (x[i]-0.5)**2 + (y[i]-0.5)**2
    d.append(1 if d_i <= 0.25 else 0)
    c.append('orange' if d_i <= 0.25 else 'pink')
pi = 4*d.count(1)/len(d)
plt.scatter(x,y,c=c)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
print(pi)
```
