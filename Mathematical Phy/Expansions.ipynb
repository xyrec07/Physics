```python
import math
import numpy as np
import matplotlib.pyplot as plt

def e_x(x):
    expansion = 0
    for i in range(100):
        expansion += x**i/math.factorial(i)
    return expansion
x_list = np.arange(-3,3,0.01)
y_list = [e_x(x) for x in x_list]
plt.plot(x_list,y_list,label='e^x')
#plt.xticks(np.arange(-10,10))
plt.grid(True)
plt.legend()
plt.show()
```

```python
def sin(x):
    expansion = 0
    for i in range(1,100,2):
        sign = 1 if i in range(1,100,4) else -1
        expansion += sign*(x**i/math.factorial(i))
    return expansion

x_list = np.arange(-5,5,0.01)
y_list = [sin(x) for x in x_list]
plt.plot(x_list,y_list,label='sin(x)')
#plt.xticks(np.arange(-10,10))
plt.grid(True)
plt.legend()
plt.show()
```

```python
def cos(x):
    expansion = 0
    for i in range(0,100,2):
        sign = 1 if i in range(0,100,4) else -1
        expansion += sign*(x**i/math.factorial(i))
    return expansion

x_list = np.arange(-5,5,0.01)
y_list = [cos(x) for x in x_list]
plt.plot(x_list,y_list,label='cos(x)')
#plt.xticks(np.arange(-10,10))
plt.grid(True)
plt.legend()
plt.show()
```

```python
def cosh(x):
    return (e_x(x)+e_x(-x))/2

x_list = np.arange(-3,3,0.01)
y_list = [cosh(x) for x in x_list]
plt.plot(x_list,y_list,label='cosh(x)')
#plt.xticks(np.arange(-10,10))
plt.grid(True)
plt.legend()
plt.show()
```

```python
def sinh(x):
    return (e_x(x)-e_x(-x))/2

x_list = np.arange(-3,3,0.01)
y_list = [sinh(x) for x in x_list]
plt.plot(x_list,y_list,label='sinh(x)')
#plt.xticks(np.arange(-10,10))
plt.grid(True)
plt.legend()
plt.show()
```
